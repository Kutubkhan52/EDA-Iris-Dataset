```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

iris1 = pd.read_csv('iris1.csv')
iris1
```

```python
iris1.plot(kind ='scatter', x ='sepal_length', y='sepal_width')

plt.grid()

plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris1, hue='species', size=4).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris1, hue='species', size =3)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris1, hue='species', size=5).map(sns.distplot, 'petal_length')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris1, hue='species', size=5).map(sns.distplot, 'petalwidth')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris1, hue='species', size=5).map(sns.distplot, 'sepal_length')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris1, hue='species', size=5).map(sns.distplot, 'sepal_width')
plt.show()
```
